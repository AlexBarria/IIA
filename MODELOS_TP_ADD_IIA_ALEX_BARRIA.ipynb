```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
#pd.options.plotting.backend = "plotly"

from sklearn import preprocessing
import seaborn as sns
from datetime import datetime
```

# CLASIFICACIÓN

```python
train_set = pd.read_pickle('./classifier_train_set.pkl')
test_set = pd.read_pickle('./classifier_test_set.pkl')
```

```python
test_set
```

```python
X_train = train_set.drop(columns = ['RainfallTomorrow', 'RainTomorrow']).values
y_train = train_set['RainTomorrow'].values

X_test = test_set.drop(columns = ['RainfallTomorrow', 'RainTomorrow']).values
y_test = test_set['RainTomorrow'].values
```

# LogisticRegression()

```python
# Defíno métricas
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
def get_classifier_scores(DF, model_name, penalty_name, C_value, class_weight, y_true, y_pred):
    rec = recall_score(y_true = y_test, y_pred = y_test_pred)
    f1 = f1_score(y_true = y_test, y_pred = y_test_pred)
    acc = accuracy_score(y_true = y_test, y_pred = y_test_pred)
    prec = precision_score(y_true = y_test, y_pred = y_test_pred)
    print('Métricas | recall: '+str(rec)+' | f1_score: '+str(f1)+' | accuracy: '+str(acc)+' | precision: '+str(prec))
    DF.append({'Model':model_name, 'Recall':rec, 'f1':f1, 'accuracy':acc,'precision':prec}, ignore_index=True)
    return DF.append({'Model':model_name, 'Penalty': penalty_name, 'C' : C_value, 'Class_weight' : class_weight, 'Recall':rec, 'f1':f1, 'accuracy':acc,'precision':prec}, ignore_index=True)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_names = [
    'Sklearn | LogisticRegression', 
    'Sklearn | LogisticRegression',
    'Sklearn | LogisticRegression',
    'Sklearn | LogisticRegression',
    'Sklearn | LogisticRegression',
    'Sklearn | LogisticRegression'
    ]

penalty_name = [
    'none',
    'none',
    'l2',
    'l2',
    'l2',
    'l2'
    ]

class_weight = [
    'none',
    'none',
    'balanced',
    'balanced',
    'balanced',
    'balanced'
]

C_value = [
    1,
    1,
    1,
    1,
    0.001,
    10
]
```

```python
model_name = 'Sklearn | LogisticRegression C'
penalty_name = 'l2'
class_weight = 'balanced'
```

```python
# BUSCO LO MEJORES PARÁMETROS UTILIZANDO GRIDSEARCH
from sklearn.model_selection import GridSearchCV
clf_grid = LogisticRegression()
grid_values = {'penalty': ['l1', 'l2', 'elasticnet'],'C':[0.001,.009,0.01,.09,1,5,10,25], 'class_weight': ['none', 'balanced']}
grid_clf_acc = GridSearchCV(clf_grid, param_grid = grid_values, scoring = 'recall')
grid_clf_acc.fit(X_train, y_train)


y_pred_acc = grid_clf_acc.predict(X_test)
```

```python
grid_clf_acc.best_params_
```

```python
# ENTRENO VARIOS MODELOS CON DIFERENTES HIPERPARÁMETROS PARA COMPARAR LOS RESULTADOS
ScoresDF = pd.DataFrame(columns = ['Model', 'Penalty', 'Class_weight', 'C', 'Recall', 'f1', 'accuracy', 'precision'])
for i in np.arange(0,len(C_value)):
    clf = LogisticRegression(max_iter = 1000, verbose = True, penalty = penalty_name[i], n_jobs = None, class_weight = class_weight[i], C = C_value[i])
    clf.fit(X_train, y_train)
    y_test_pred = clf.predict(X_test)
    ScoresDF = get_classifier_scores(ScoresDF, model_names[i], penalty_name[i], C_value[i], class_weight[i], y_test, y_test_pred)
```

```python
# RESULTADOS OBTENIDOS
ScoresDF
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
# Defíno métricas
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
ScoresDF_Regression = pd.DataFrame(columns = ['Model', 'Mean_absolute_error', 'Mean_squared_error', 'r2_score'])
```

```python
def get_regressor_scores(DF, model_name, y_true, y_pred):
    mae = mean_absolute_error(y_true = y_test, y_pred = y_test_pred)
    mse = mean_squared_error(y_true = y_test, y_pred = y_test_pred)
    r2 = r2_score(y_true = y_test, y_pred = y_test_pred)
    
    print('Métricas | MAE: '+str(mae)+' | MSE: '+str(mse)+' | R2: '+str(r2))
    DF.append({'Model':model_name, 'Mean_absolute_error':mae, 'Mean_squared_error':mse, 'r2_score':r2}, ignore_index=True)
    return DF.append({'Model':model_name, 'Mean_absolute_error':mae, 'Mean_squared_error':mse, 'r2_score':r2}, ignore_index=True)
```

```python
LinearRegressor = LinearRegression()
```

```python
LinearRegressor.fit(X_train, y_train)
```

```python
y_test_pred = LinearRegressor.predict(X_test)
```

```python
model_name = 'LinearRegressor'
ScoresDF_Regression = get_regressor_scores(ScoresDF_Regression, model_name, y_test, y_test_pred)
```
